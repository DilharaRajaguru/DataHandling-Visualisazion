```python
import pandas as pd
#import io
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

df = pd.read_csv('data6-1.csv')
# print(df)
data = np.array(df)
print(data)

```

```python
def  probability_density(data):
  # plt.hist(data)
  plt.hist(data, bins=30,  alpha=0.75, edgecolor='black', label='Salary Distribution')
  plt.show()

probability_density(data)
```

```python
#mean salary W~
mean_salary = np.mean(data)
print(f'Mean Annual Salary (W˜): {mean_salary}')

#mean salary 1.25W~
new_mean_salary = 1.25 * mean_salary
print(new_mean_salary)
```

```python
#standard deviation

sigma = np.std(data, ddof=1)
print(sigma)
```

```python
W = mean_salary
z_w = (W - mean_salary) / sigma
z_1_25w = (1.25 * W - mean_salary) / sigma
print(z_w)
print(z_1_25w)
```

```python
# Calculate the cumulative probabilities
cdf_w = stats.norm.cdf(z_w)
cdf_1_25w = stats.norm.cdf(z_1_25w)

# Calculate the fraction of the population between W˜ and 1.25W˜
fraction_population = cdf_1_25w - cdf_w
#rounded_result = round(fraction_population , 2)
print(fraction_population)
#print(rounded_result)
```

```python
# Plotting the histogram with labels, title, and legend
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.75, color='b', edgecolor='black', label='Salary Distribution')
plt.axvline(mean_salary, color='r', linestyle='dashed', linewidth=2, label=f'Mean Salary ($W˜$): {round(mean_salary, 2)}')
plt.axvline(new_mean_salary, color='g', linestyle='dashed', linewidth=2, label=f'Mean Salary ($1.25W˜$): {round(new_mean_salary, 2)}')
#plt.axvline(fraction_population, color='b', linestyle='dashed', linewidth=2, label=f'Fraction Value (X): {round(fraction_population, 2)}')

plt.legend( title=f'Fraction (X): {round(fraction_population, 2)}')
plt.xlabel('Salary')
plt.ylabel('Probability Density')
plt.title('Salary Distribution and Mean Salaries')

# Print fraction of population with salaries between W˜ and 1.25W˜ (X) on the graph
# plt.text(0.8, 0.9, f'Fraction (X): {round(fraction_population, 4)}', transform=plt.gca().transAxes, fontsize=10, verticalalignment='top')

# Save the plot to a file
plt.savefig('salary_distribution.png')

# Show the plot
plt.show()
```
